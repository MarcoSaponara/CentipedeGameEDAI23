```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
icg4_sd = np.load('./icg4_sd.npy')
icg4_js = np.load('./icg4_js.npy')
icg4_betas = np.load('./icg4_beta_array.npy')
icg4_rmse = np.load('./icg4_rmse.npy')
icg4_hel = np.load('./icg4_hel.npy')
icg4_eps = .09

icg4_idx_beta_best_fit_rmse = np.argmin(icg4_rmse)
icg4_beta_best_fit_rmse = icg4_betas[icg4_idx_beta_best_fit_rmse]

icg4_idx_beta_best_fit_js = np.argmin(icg4_js)
icg4_beta_best_fit_js = icg4_betas[icg4_idx_beta_best_fit_js]

icg4_idx_beta_best_fit_hel = np.argmin(icg4_hel)
icg4_beta_best_fit_hel = icg4_betas[icg4_idx_beta_best_fit_hel]
```

```python
icg6_sd = np.load('./icg6_sd.npy')
icg6_js = np.load('./icg6_js.npy')
icg6_betas = np.load('./icg6_beta_array.npy')
icg6_rmse = np.load('./icg6_rmse.npy')
icg6_hel = np.load('./icg6_hel.npy')
icg6_eps = .045

icg6_idx_beta_best_fit_rmse = np.argmin(icg6_rmse)
icg6_beta_best_fit_rmse = icg6_betas[icg6_idx_beta_best_fit_rmse]

icg6_idx_beta_best_fit_js = np.argmin(icg6_js)
icg6_beta_best_fit_js = icg6_betas[icg6_idx_beta_best_fit_js]

icg6_idx_beta_best_fit_hel = np.argmin(icg6_hel)
icg6_beta_best_fit_hel = icg6_betas[icg6_idx_beta_best_fit_hel]
```

```python
figure, axis = plt.subplots(2, 2, figsize=(14,7), dpi=300, sharex=True)

axis[0,0].plot(icg4_betas, icg4_rmse, color='green',)
axis[0,0].plot(icg4_betas, icg4_js, color='orange',)
axis[0,0].plot(icg4_betas, icg4_hel, color='purple',)

axis[0,0].axvline(x=icg4_beta_best_fit_rmse, color='green', alpha=.75, linestyle='--')
axis[0,0].axvline(x=icg4_beta_best_fit_js, color='orange', alpha=.75, linestyle='-.')
axis[0,0].axvline(x=icg4_beta_best_fit_hel, color='purple', alpha=.75, linestyle=':')

axis[0,0].axhline(y=np.min(icg4_rmse), color='green', alpha=.75, linestyle='--')
axis[0,0].axhline(y=np.min(icg4_js), color='orange', alpha=.75, linestyle='-.')
axis[0,0].axhline(y=np.min(icg4_hel), color='purple', alpha=.75, linestyle=':')

axis[0,0].set_title(r'$L=4$, $\epsilon=$'+str(icg4_eps)+'\nQuality of fit (model vs experiment)')
axis[0,0].set_ylabel('distance between model & experiment')
axis[0,0].set_xlabel(r'$\beta$ (intensity of selection)')
axis[0,0].set_xscale('log')
axis[0,0].set_ylim(-.01, 0.75)
axis[0,0].legend(['RMSE', 'Jensen-Shannon', 'Hellinger'], loc='upper left')

###

axis[0,1].plot(icg6_betas, icg6_rmse, color='green',)
axis[0,1].plot(icg6_betas, icg6_js, color='orange',)
axis[0,1].plot(icg6_betas, icg6_hel, color='purple',)

axis[0,1].axvline(x=icg6_beta_best_fit_rmse, color='green', alpha=.75, linestyle='--')
axis[0,1].axvline(x=icg6_beta_best_fit_js, color='orange', alpha=.75, linestyle='-.')
axis[0,1].axvline(x=icg6_beta_best_fit_hel, color='purple', alpha=.75, linestyle=':')

axis[0,1].axhline(y=np.min(icg6_rmse), color='green', alpha=.75, linestyle='--')
axis[0,1].axhline(y=np.min(icg6_js), color='orange', alpha=.75, linestyle='-.')
axis[0,1].axhline(y=np.min(icg6_hel), color='purple', alpha=.75, linestyle=':')

axis[0,1].set_title(r'$L=6$, $\epsilon=$'+str(icg6_eps)+'\nQuality of fit (model vs experiment)')
axis[0,1].set_ylabel('distance between model & experiment')
axis[0,1].set_xlabel(r'$\beta$ (intensity of selection)')
axis[0,1].set_xscale('log')
axis[0,1].set_ylim(-.01, 0.75)
axis[0,1].legend(['RMSE', 'Jensen-Shannon', 'Hellinger'], loc='upper left')

###

axis[1,0].plot(icg4_betas, icg4_sd)
axis[1,0].axvline(x=icg4_beta_best_fit_rmse, color='black', alpha=.75, linestyle='--')
axis[1,0].axvline(x=icg4_beta_best_fit_js, color='black', alpha=.75, linestyle='-.')
axis[1,0].axvline(x=icg4_beta_best_fit_hel, color='black', alpha=.75, linestyle=':')
axis[1,0].set_title(r'Distribution of strategies for different values of $\beta$')
axis[1,0].set_xlabel(r'$\beta$ (intensity of selection)')
axis[1,0].set_ylabel('frequency')
axis[1,0].set_xscale('log')
axis[1,0].set_ylim(-.01, 0.8)
axis[1,0].legend(['k='+str(t) for t in range(5)], loc='upper left')

###
 
axis[1,1].plot(icg6_betas, icg6_sd)
axis[1,1].axvline(x=icg6_beta_best_fit_rmse, color='black', alpha=.75, linestyle='--')
axis[1,1].axvline(x=icg6_beta_best_fit_js, color='black', alpha=.75, linestyle='-.')
axis[1,1].axvline(x=icg6_beta_best_fit_hel, color='black', alpha=.75, linestyle=':')
axis[1,1].set_title(r'Distribution of strategies for different values of $\beta$')
axis[1,1].set_xlabel(r'$\beta$ (intensity of selection)')
axis[1,1].set_ylabel('frequency')
axis[1,1].set_xscale('log')
axis[1,1].set_ylim(-.01, 0.8)
axis[1,1].legend(['k='+str(t) for t in range(7)], loc='upper left')


figure.tight_layout() 
plt.savefig('./fig2.png',format='png')
plt.savefig('./fig2_vec.svg',format='svg')
plt.show()
```
