```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
icg4_js = np.load('./icg4_js_grid.npy')
icg4_betas = np.load('./icg4_beta_array.npy')
icg4_epsilons = np.load('./icg4_eps_array.npy')
```

```python
icg6_js = np.load('./icg6_js_grid.npy')
icg6_betas = np.load('./icg6_beta_array.npy')
icg6_epsilons = np.load('./icg6_eps_array.npy')
```

```python
vmin = np.min([np.min(np.log10(icg4_js)), np.min(np.log10(icg6_js))])
vmax = np.max([np.max(np.log10(icg4_js)), np.max(np.log10(icg6_js))])

fig, ax = plt.subplots(2, 1, figsize=(5, 8), dpi=300, sharex=True, sharey=True)
#fig.suptitle('Distance between Take distributions (model vs experiment)')

tmp = ax[0].pcolor(icg4_betas, icg4_epsilons, np.log10(icg4_js), cmap='viridis', vmin = vmin, vmax = vmax)
for i, beta in enumerate(icg4_betas):
    ax[0].scatter(x=beta, y=icg4_epsilons[np.argmin(icg4_js[:, i])], c='white',s = 2)
ax[0].set_xscale('log')
# ax[0].set_yscale('log')
ax[0].set_xlabel(r'$\beta$ (intensity of selection)')
ax[0].set_ylabel(r'$\epsilon$ (reasoning error probability)')
ax[0].set_title(r'$L=4$')

tmp = ax[1].pcolor(icg6_betas, icg6_epsilons, np.log10(icg6_js), cmap='viridis', vmin = vmin, vmax = vmax)
for i, beta in enumerate(icg6_betas):
    ax[1].scatter(x=beta, y=icg6_epsilons[np.argmin(icg6_js[:, i])], c='white',s = 2)
ax[1].set_xscale('log')
# ax[1].set_yscale('log')
ax[1].set_xlabel(r'$\beta$ (intensity of selection)')
ax[1].set_ylabel(r'$\epsilon$ (reasoning error probability)')
ax[1].set_title(r'$L=6$')

fig.colorbar(tmp, fraction=0.05,ax=ax.ravel().tolist(), )# orientation='horizontal')
# fig.tight_layout()
plt.savefig('./fig1.png',format='png')
plt.savefig('./fig1_vec.svg',format='svg')
plt.show() # display
```
